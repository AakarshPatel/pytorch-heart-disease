### Initial Observations from Summary Statistics
- Age ranges from 29 to 77 years, with most patients in middle age (mean ~54).
- **Sex:** Skewed toward male patients (~68% male).
- **Chest pain type (`cp`):** Values 14 indicate different types of pain; majority are type 3
- **Resting blood pressure (`trestbps`):** 94200 mm Hg, mean ~132 mm Hg. A few high values may be potential outliers.
-**Cholesterol (`chol`):** 126564 mg/dl, mean ~247 mg/dl. Some very high readings suggest potential outliers.  
- **Fasting blood sugar (`fbs`):** ~15% of patients have high fasting blood sugar (>120 mg/dl).  
- **Resting ECG (`restecg`):** Most values are 0 or 1, representing normal vs minor abnormalities.  
- **Maximum heart rate achieved (`thalach`):** 71202 bpm, mean ~150 bpm.  
- **Exercise-induced angina (`exang`):** ~33% of patients show angina during exercise.  
- **ST depression (`oldpeak`):** 06.2; higher values indicate greater exercise-induced depression, possible outliers.  
- **Slope of peak ST segment (`slope`):** 13, showing upsloping, flat, or downsloping; mostly 1 or 2.  
- **Number of major vessels colored (`ca`):** Contains `?` values; currently treated as object type and will be cleaned later.  
- **Thalassemia (`thal`):** Contains `?` values; to be handled in data cleaning.  
- **Target (`num`):** Min = 0, max = 4, mean ~0.94. Will be binarized for heart disease presence vs absence. Slight class imbalance exists (more patients without disease).

**Notes:**  
- Columns `ca` and `thal` contain missing entries represented as `?`.  
- Outliers in `chol`, `trestbps`, and `oldpeak` should be noted for later analysis.  
- Using only the 14 relevant features aligns with ML research on the Cleveland dataset.



```python
count0 = 0
count1 = 0
count2 = 0
count3 = 0
count4 = 0
for index, row_data in data.iterrows(): # Iterate over rows to count occurrences of each 'num' value.
  value = row_data['num'] # Access 'num' column from the current row Series
  if value == 0:
    count0 += 1
  elif value == 1:
    count1 += 1
  elif value == 2:
    count2 += 1
  elif value == 3:
    count3 += 1
  else:
    count4 += 1
print(f"num people with presence 0: {count0} \nnum people with presence 1: {count1} \nnum people with presence 2: {count2} \nnum people with presence 3: {count3} \nnum people with presence 4: {count4}") # Print the counts for each 'num' category.
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
sns.histplot(data['num'], kde=False, bins=15)
plt.title('Distribution of Values Column (Seaborn)')
plt.xlabel('Value')
plt.ylabel('num')
plt.xticks(ticks=[0,1,2,3,4])
plt.show()
```

### Graph 1: Distribution of Heart Disease Presence (`num`)

This histogram shows the distribution of the `num` column, which indicates the presence of heart disease (0 = no disease, 1-4 = various levels of disease presence). It clearly highlights that a significant portion of the dataset corresponds to `num=0` (no disease). The counts then decrease for `num=1`, `num=2`, `num=3`, and `num=4`.

**Comments:**
*   This visualization confirms the need to handle potential class imbalance if `num` is to be used as a multi-class target variable directly in a machine learning model.
*   For binary classification, the `num` column would typically be binarized (0 for no disease, 1 for any presence of disease).
